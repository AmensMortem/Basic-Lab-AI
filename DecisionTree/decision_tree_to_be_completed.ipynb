n# Decision Trees

A **Decision Tree Classifier** is a supervised learning algorithm that assigns class labels to instances by recursively partitioning the feature space according to data-driven decision rules.

At each internal node, the algorithm evaluates a condition on a specific feature and splits the data into subsets based on the outcome of that condition. The branches represent these outcomes, while the leaf nodes correspond to the final predicted class labels.

This hierarchical, tree-structured representation enables the model to capture non-linear decision boundaries while maintaining a high level of interpretability, making it particularly suitable for classification tasks.




---
## Structure

It has a hierarchical, tree structure, which consists of a root node, branches, internal nodes and leaf nodes.
- **Root node** represents one feature of the data.
- **Internal nodes** represent features (or attributes) of the data.
- **Branches** represent the decision rules.
- **Leaf nodes** represent the outcome.



Each internal node performs a test like:

```
Is Age > 30?
- No: Go Left
- Yes: Go Right
```

Based on the answer (Yes/No), the data follows a specific branch.

---

## How Decision Trees Work - Overview

The algorithm follows these main steps:

1. Start with the entire dataset.
2. Select the best feature to split the data.
3. Partition the dataset into subsets.
4. Repeat the process recursively on each subset.
5. Stop when:

   * All samples belong to the same class, or
   * A stopping criterion is reached (max depth, minimum samples, etc.).

This recursive partitioning process is called **tree induction**.

---

## Splitting Criteria (How the Best Feature is Chosen)

For **classification**, common metrics are:

* **Gini Index**
* **Entropy (Information Gain)** (Better one)

### Gini Index

$
Gini = 1 - \sum_{i=1}^{C} p_i^2
$

Where ( $p_i$ ) is the probability of class ( i ).

Lower Gini → Better split.

---

### Entropy

$
Entropy = -\sum_{i=1}^{C} p_i \log_2(p_i)
$

Information Gain:

$
IG = Entropy(parent) - \sum Weighted\ Entropy(children)
$

Higher Information Gain → Better split.

---

For **regression**, the common criterion is:

* **Mean Squared Error (MSE)**

$
MSE = \frac{1}{n} \sum (y_i - \hat{y})^2
$

The algorithm selects splits that minimize variance.

---

In a **Decision Tree Classifier**, the probability of a class is computed from the **distribution of training samples that reach a leaf node**. The tree itself does not explicitly learn probabilities; instead, it estimates them using **class frequencies** in the leaves.

#### Probability at a Leaf Node

Suppose a leaf node contains (N) training samples.

Let:

* ($N_c$) = number of samples belonging to class (c)
* (N) = total samples in the leaf

The probability of class (c) is:

$
P(c) = \frac{N_c}{N}
$

This means the probability is simply the **relative frequency of that class in the leaf**.

---

#### Example

Consider a leaf node containing 10 training samples:

| Class   | Count |
| ------- | ----- |
| Class A | 7     |
| Class B | 3     |

Total samples:

$
N = 10
$

### Probability of each class

$
P(A) = \frac{7}{10} = 0.7
$

$
P(B) = \frac{3}{10} = 0.3
$

---
#### Treating the Data

````python
X = diabetes_data.drop("Outcome", axis=1)
y = diabetes_data["Outcome"]

X.shape, y.shape
````

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42
)
````

````python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
````

---
#### Decision tree and Random Search

````python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV
````

````python
tree = DecisionTreeClassifier(random_state=0)
````

````python
distributions = {
    "criterion": ['entropy', "gini"],
    "min_samples_split": range(2,11,1),
    "min_samples_leaf": range(2,11,1),
}
````

````python
clf = RandomizedSearchCV(
    estimator=tree,
    param_distributions=distributions,
    cv=5,
    random_state=0,
    n_jobs=1,
    scoring = 'accuracy'
)
````

````python
search_data = clf.fit(X_train_scaled, y_train)
````

````python
search_data.best_params_
````

---
#### Using the Best hyperparameters

````python
tree = DecisionTreeClassifier(**search_data.best_params_)
````

````python
tree.fit(X_train_scaled, y_train)
````

````python
y_pred = tree.predict(X_test_scaled)
````

````python
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score
````

````python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test.to_numpy(), y_pred, average="binary")
recall = recall_score(y_test, y_pred, average="binary")

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
````

````python
print(classification_report(y_test,y_pred))
````
